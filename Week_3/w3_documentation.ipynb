# Student Average Calculator Documentation

## 1. Identify the Components

### 1.1 What are the inputs?

The inputs are the quiz marks entered by the student.

The program requires three inputs:
- Quiz 1 mark (float)
- Quiz 2 mark (float)
- Quiz 3 mark (float)


### 1.2 What is the process?

The process is:
1. Get the quiz marks from the user.
2. Calculate the total marks.
3. Calculate the average mark.
4. Check whether the student passes or fails.


### 1.3 What is the output?

The outputs are:
1. The student's average mark.
2. The result of the student (Pass or Fail).

# Student Average Calculator

The program will:
1. Take the marks from the user.

The program will show:
- The average mark.
- Pass or Fail result.


# 2. Design the Algorithm

## 2.1 Pseudocode

Start

Set choice to "y"

While choice is "y":

    Enter three quiz marks

    Calculate average

    Display average

    If average is 50 or above:
        Print Pass
    Otherwise:
        Print Fail

    Ask user if they want to continue

End while

Print "Program Ended"

End


## 2.2 Algorithm Structure

### Sequence:
The program takes input, calculates the average, and displays the result.

### Selection:
The program uses if-else to check Pass or Fail.

### Iteration:
The program uses a while loop to repeat the program.
